```python
import tensorflow as tf
from tensorflow.keras import layers
from tensorflow.keras.preprocessing.image import img_to_array, load_img
import numpy as np
from google.colab import files
import io
```

```python
safePhotos = files.upload()
brokenPhotos = files.upload()
```

```python
safeImages = [img_to_array(load_img(io.BytesIO(safePhotos[key]), target_size=(512, 512))) for key in safePhotos.keys()]
brokenImages = [img_to_array(load_img(io.BytesIO(brokenPhotos[key]), target_size=(512, 512))) for key in brokenPhotos.keys()]
```

```python
safeLabels = np.ones(len(safeImages))
brokenLabels = np.zeros(len(brokenImages))
```

```python
allImages = np.concatenate((safeImages, brokenImages), axis=0)
allLabels = np.concatenate((safeLabels, brokenLabels), axis=0)
```

```python
indices = np.arange(allImages.shape[0])
np.random.shuffle(indices)
allImages = allImages[indices]
allLabels = allLabels[indices]
```

```python
dataset = tf.data.Dataset.from_tensor_slices((allImages, allLabels))
```

```python
# Define the model
model = tf.keras.Sequential([
    layers.InputLayer(input_shape=(512, 512, 3)),
    layers.Conv2D(16, 3, activation='relu'),
    layers.MaxPooling2D(),
    layers.Conv2D(32, 3, activation='relu'),
    layers.MaxPooling2D(),
    layers.Conv2D(64, 3, activation='relu'),
    layers.MaxPooling2D(),
    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dense(1, activation='sigmoid')
])
```

```python
# Compile the model
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python
# Train the model
batch_size = 32
epochs = 10
model.fit(dataset.batch(batch_size), epochs=epochs)
```

```python
newImage = files.upload()
newImageArray = [img_to_array(load_img(io.BytesIO(newImage[key]), target_size=(512, 512))) for key in newImage.keys()]
newImageArray = np.array(newImageArray)

with open('predictions_output.txt', 'w') as file:
  predictions = model.predict(newImageArray)
  for index, prediction in enumerate(predictions):
      if prediction[0] >= 0.5:
          result = (f"Position{index + 1} Safe\n")
          print(result)

      else:
          result = (f"Position{index + 1} AtRisk\n")
          print(result)

      file.write(result)

files.download('predictions_output.txt')
```
